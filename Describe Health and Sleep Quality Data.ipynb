#### Dataset Overview
The Sleep Health and Lifestyle Dataset comprises a wide range of variables related to sleep and daily habits. It includes details such as gender, age, occupation, sleep duration, quality of sleep, physical activity level, stress levels, BMI category, blood pressure, heart rate, daily steps, and the presence or absence of sleep disorders.

#### Dataset Columns: 

Person ID: An identifier for each individual.

Gender: The gender of the person (Male/Female).

Age: The age of the person in years.

Occupation: The occupation or profession of the person.

Sleep Duration (hours): The number of hours the person sleeps per day.

Quality of Sleep (scale: 1-10): A subjective rating of the quality of sleep, ranging from 1 to 10.

Physical Activity Level (minutes/day): The number of minutes the person engages in physical activity daily.

Stress Level (scale: 1-10): A subjective rating of the stress level experienced by the person, ranging from 1 to 10.

BMI Category: The BMI category of the person (e.g., Underweight, Normal, Overweight).

Blood Pressure (systolic/diastolic): The blood pressure measurement of the person, indicated as systolic pressure over diastolic pressure.

Heart Rate (bpm): The resting heart rate of the person in beats per minute.

Daily Steps: The number of steps the person takes per day.

Sleep Disorder: The presence or absence of a sleep disorder in the person (None, Insomnia, Sleep Apnea).

#### Details about Sleep Disorder Column: 

None: The individual does not exhibit any specific sleep disorder.
Insomnia: The individual experiences difficulty falling asleep or staying asleep, leading to inadequate or poor-quality sleep.
Sleep Apnea: The individual suffers from pauses in breathing during sleep, resulting in disrupted sleep patterns and potential health risks.

## 3 - Analysis of Daily Steps Taken

#### Basic statistics of 'Daily Steps'

```python
lifestyle_df['Daily Steps'].describe()
```

```python
daily_steps = lifestyle_df['Daily Steps']

std_val = daily_steps.std()
max_val = daily_steps.max()
min_val = daily_steps.min()
range_val = max_val - min_val

print('Standard Deviation = ', std_val)
print('Maximum = ', max_val)
print('Minumum = ', min_val)
print('Range = ', range_val)
```

#### Answer-set-3 [Typical Amount (Minutes) of Physical Activity]
    3a - On this slide, provide the standard deviation, maximum, minimum, and range associated with the number of steps taken.
       - Standard Deviation = 1617.915679133637
       - Maximum = 10000
       - Minumum = 3000
       - Range = 7000
       - The large range and standard deviation indicate substantial variability in the number of daily steps among participants.
    3b - We can use IQR and the boxplot. 

```python
q1 = lifestyle_df['Daily Steps'].quantile(0.25)
q3 = lifestyle_df['Daily Steps'].quantile(0.75)
iqr = q3 - q1

print("Q1:", q1)
print("Q3:", q3)
print("IQR:", iqr)
```

#### Note: The interquartile range (IQR) measures the spread of the middle 50% of the data and is less affected by outliers than the overall range.

```python
fig, ax = plt.subplots(figsize=(10,5))

lifestyle_df.boxplot(column='Daily Steps', vert=False, ax=ax)
utl.x_y_t(ax, 'Daily Steps', '', 'Boxplot of Daily Steps')

plt.show()
```

#### Note: In addition to the standard deviation and range, the interquartile range (IQR) was calculated to better understand the spread of the middle 50% of the data. The IQR is less sensitive to extreme values and provides a robust measure of variability.

## 4 - Distribution of Heart Rates

#### Basic statistics of 'Heart Rates'

```python
lifestyle_df['Heart Rate'].describe()
```

#### Note: The average heart rate is approximately 70 bpm with relatively low variability (std ≈ 4.1). The median is close to the mean, suggesting the distribution is approximately symmetric with no major skewness. We will check both histogram and boxplot. 

```python
heart_rate = lifestyle_df['Heart Rate']

print("Mean:", heart_rate.mean())
print("Median:", heart_rate.median())
print("Skew:", heart_rate.skew())
```

```python
fig, ax = plt.subplots(figsize=(10,5))

lifestyle_df['Heart Rate'].plot(kind='hist', bins=10, ax=ax)
utl.x_y_t(ax,
          'Heart Rate',
          'Frequency',
          'Distribution of Heart Rates')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,5))

lifestyle_df.boxplot(column='Heart Rate', vert=False, ax=ax)
utl.x_y_t(ax,
          'Heart Rate',
          '',
          'Boxplot of Heart Rate')

plt.show()
```

#### Answer-set-4 [Distribution of Heart Rates]
    4a - The histogram of heart rates appears approximately symmetric and bell-shaped, indicating a near-normal distribution. The boxplot shows moderate spread with no significant outliers.
